# House pricing programming assignment

We will work with a benchmark dataset for houses prices that contains both visual and textual information. Each house is represented by four images for bedroom, bathroom, kitchen and a frontal image of the house. The data folder contains 2140 images, 4 images for each house. Also, it contains a text file that contains the textual metadata of the dataset. Each row in the file represents the number of house in order. The numbers represent number of bedrooms, number of bathrooms, area of the house, zipcode and the price.

**Number of Instances:** 535

**Number of Attributes:** 4 textual attributes in addition to the visual attributes that can be extracted from the images. 

**Attribute Information:**
1. Number of Bedrooms      
2. Number of bathrooms   
3. Area
4. Zipcode
5. Price


For more info and credits visit: [https://github.com/omarsayed7/House-price-estimation-from-visual-and-textual-features](https://github.com/omarsayed7/House-price-estimation-from-visual-and-textual-features)
